```python
def processDate(record):
    
    riskFree = record['Shibor 3M'] / 100.0
    maturity = record['Maturity']
    spot = record['161207.OFCN']
    ATarget = record['150008.XSHE']
    BTarget = record['150009.XSHE']
    
    option1Strike = record['Strike 1']
    option1Type =  record['Type'] == 'Call' and 1 or -1
    option1A = record['A']
    option1B = record['B']
    option1Direction = record['direction']
    
    option2Strike = record['Strike 2']
    option2Type =  record['Type.1'] == 'Call' and 1 or -1
    option2A = record['A.1']
    option2B = record['B.1']
    option2Direction = record['direction.1']
    
    option3Strike = record['Strike 3']
    option3Type =  record['Type.2'] == 'Call' and 1 or -1
    option3A = record['A.2']
    option3B = record['B.2']
    option3Direction = record['direction.2']
    
    def ABOptionPrice(vol):
        price1 = BSMPrice(option1Type, option1Strike, spot, riskFree, 0.0, vol[0], maturity, rawOutput = True)
        price2 = BSMPrice(option2Type, option2Strike, spot, riskFree, 0.0, vol[0], maturity, rawOutput = True)
        price3 = BSMPrice(option3Type, option3Strike, spot, riskFree, 0.0, vol[0], maturity, rawOutput = True) 
        aPrice = 1.0 + option1Direction * price1[0] * option1A + option2Direction * price2[0] * option2A + option3Direction * price3[0] * option3A
        bPrice = 1.0 + option1Direction * price1[0] * option1B + option2Direction * price2[0] * option2B + option3Direction * price3[0] * option3B
        
        return aPrice -  ATarget, bPrice -BTarget
    
    def errorFunction(vol):
        
        aPrice, bPrice = ABOptionPrice(vol)
        
        return aPrice**2  + bPrice**2
        
    out, fx, its, imode, smode = optimize.fmin_slsqp(errorFunction, [0.15], bounds = [(0.001, 0.50)], epsilon = 1e-6, iter = 10000, disp = False, full_output = True, acc = 1e-16)
    
    return ABOptionPrice(out)
```

```python
for i in xrange(len(res)):
    aprice, bprice  = processDate(res.loc[i])
    print aprice, bprice
```
